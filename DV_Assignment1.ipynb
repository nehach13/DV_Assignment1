```python
import os
import calendar
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from pandas.plotting import scatter_matrix, parallel_coordinates
import seaborn as sns
from sklearn import preprocessing
%matplotlib inline
```

```python
df = pd.read_csv("ApplianceShipments.csv") 
```

```python
df.head()
```

```python
#line
quarter = df['Quarter']
shipments = df['Shipments']
plt.plot(quarter,shipments,color = 'blue')
plt.title('US household appliances sold by quarter')
plt.xticks(rotation='vertical')
plt.xlabel('Quarter')
plt.ylabel('Shipments')
plt.show()
```

The line graph shows the shipments in every quarter from year 1985 - 1989. As shown the number of shipments has increased gradually through all the years and highest being in Q2 of 1989.

```python
#barchart
quarter = df['Quarter']
shipments = df['Shipments']
plt.bar(quarter,shipments,color = 'orange')
plt.title('US household appliances sold by quarter')
plt.xticks(rotation='vertical')
plt.xlabel('Quarter')
plt.ylabel('Shipments')
plt.show()
```

The bar chart shows the appliances sold by the store in each quarter from year 1985 to 1989.  As seen in the bar chart Q2 of year 1989 has the largest sale of appliances in all the 4 years which is nearly 5000 and least sale in Q4 of 1985 which is less than 4000


```python
#piechart
quarter = df['Quarter']
shipments = df['Shipments']
plt.pie(shipments,labels=quarter,autopct='%1.1f%%',shadow=True, startangle=90, radius = 2)
plt.title('US household appliances sold by quarter')
plt.show()
```

The pie chart shows the appliances sold in each quarter from year 1985 to 1989. As seen in the pie chart Q2 of year 1989 has the largest sale of appliances in all the 4 years i.e. 5.5% of all sale and least sale in Q1 and Q4 of 1985 i.e. 4.5% of all sale

```python
df1 = pd.read_csv("RidingMowers.csv") #read csv file and save it as a dataframe
df1.head()
```

```python
#scatterplot
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(df1[df1.Ownership == 'Owner'].Lot_Size, df1[df1.Ownership == 'Owner'].Income, color = 'Orange')
ax.scatter(df1[df1.Ownership == 'Nonowner'].Lot_Size, df1[df1.Ownership == 'Nonowner'].Income,  color = 'Blue')

ax.set_xlabel("Lot Size")
ax.set_ylabel("Income")
ax.legend(["Owner", "Nonowner"])
```

This scatterplot shows that how the lot size of the property changes according to the income for both owners and non owners of the property. As seen the income is higher for the owners and hence the lot size of their property is bigger. Though some of the non owners has good income and good lot size of the property.

```python
#histogram
sales = laptops_df['Store Postcode']
plt.hist(sales, edgecolor='black', alpha = 1, bins = 10)
plt.title('Sales by Store')
plt.xticks(rotation='vertical')
plt.xlabel('Stores')
plt.ylabel('Sales')
plt.show()
```

The histogram shows the sales of laptop in each store. The highest sale of laptops has been in store E7 8NW.
